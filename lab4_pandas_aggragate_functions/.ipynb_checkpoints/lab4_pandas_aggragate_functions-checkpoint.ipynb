```python
import pandas as pd
import numpy as np

youtube = pd.read_json("YouTubeDataset_withChannelElapsed.json")

youtube
```

### Using aggregate functions

```python
# Show general description of parameters of "channelViewCount" column 
youtube.channelViewCount.describe()
```

```python
# Show unique values in "videoCategoryId" column
youtube.videoCategoryId.unique()
```

```python
# Show the "channelViewCount" column's mean
youtube.channelViewCount.mean()
```

```python
# Show unique values in "videoPublished" column, type - object

youtube.videoPublished.unique()
```

```python
# Show general description of parameters of "videoPublished" column, type - object
youtube.videoPublished.describe()
```

```python
# Show counts of values of each "channelId" 
youtube.channelId.value_counts()
```

```python
# Show all counts of particular video's likes and dislikes 
youtube.loc[:, "videoLikeCount":"videoDislikeCount"]
```

```python
# Show means of all videos' likes and dislikes 
youtube.loc[:, "videoLikeCount":"videoDislikeCount"].mean()
```

```python
# Decrease all videos dislike counts by their average value of all the column
youtube_video_dislikes_mean = youtube.videoDislikeCount.mean()
youtube.videoDislikeCount.map(lambda count: count - youtube_video_dislikes_mean)
```

```python
# Add some minor changes to all videos like counts
youtube_video_likes_mean = youtube.videoLikeCount.mean()

def remean_likes(row):
    row.videoLikeCount = row.videoLikeCount - youtube_video_likes_mean
    row.videoDislikeCount = row.videoDislikeCount + row.videoDislikeCount*0.5
    return row

youtube.apply(remean_likes, axis='columns')
```

```python
youtube.head()
```

```python
# Decrease all like counts by average ratio of likes and dislikes
youtube_video_likes_to_dis_mean = youtube.videoLikeCount.mean() / youtube.videoDislikeCount.mean()
youtube.videoLikeCount - youtube_video_likes_to_dis_mean
```

```python
# Because some of the column names contain a divider in them, we must use constructions like the following:
youtube["likes/subscriber"].astype(str) + " - " + youtube["views/subscribers"].astype(str)
```

```python
# Aslo we have to manualy convert numeric types to string ones before combined outputing them using string builder
youtube.channelViewCount.astype(str) + " - " + youtube.videoCount.astype(str)
```
