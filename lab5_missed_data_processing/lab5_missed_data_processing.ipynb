# Lab work 5. Missed data processing
### A simpele dataset "House Prices" with specially omitted data was selected for this work from the following Kaggle similar [competition](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/overview). Both train.csv and test.csv are suitable for us, nevertheless for our purposes it's better to concat these two small datasets to get one bigger.

***Links to download: [train.cvs](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data?select=train.csv), [test.csv](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data?select=test.csv)***

```python
# Modules impoting
import pandas as pd
import numpy as np

# Reading the first part of our dataset
hp_data1 = pd.read_csv("train.csv")
hp_data1
```

```python
# Reading the second part of our dataset
hp_data2 = pd.read_csv("test.csv")
hp_data2
```

```python
# Combining two readed parts into one final dataset
hp_data = pd.concat([hp_data1, hp_data2])
hp_data
```

```python
# Setting up a seed of the build-in random generator
np.random.seed(0) 
```

###  The first acquaintance with the data

```python
# Printing first five rows of readed dataset
# Even now we can mention some missed data
hp_data.head()
```

### Determining the amount of missing data

```python
#Getting the amount of missed elements in all columns
missing_values_count = hp_data.isnull().sum()

# Printin the amount of missed elements in first twenty columns
missing_values_count[0:20]
```

```python
# Getting a general amount of missed elements of the data
total_missing = missing_values_count.sum()

# Getting a percent of missed elements of the data
total_cells = np.product(hp_data.shape)
percent_missing = (total_missing/total_cells) * 100
print(percent_missing, "%")
```

### Determining the reason of missing data

```python
missing_values_count[3:7:3]
```

### Missed elements deleting

```python
# Deleting all rows that contain missed element
hp_data.dropna()
```

```python
# Deleting all columns that contain missed element
columns_with_na_dropped = hp_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# Determining the amounth of rows and columns that conatins missed data
print("Total rows: %d \n" % hp_data.shape[1])
print("Rows with missed elements: %d" % columns_with_na_dropped.shape[1])
```

### Filling all missed elements with zeros

```python
hp_data.fillna(0)
```

### Making a sub-dataset with two most unfilled columns

```python
subset_hp_data = hp_data[['LotFrontage', 'Alley']]
subset_hp_data
```

```python
# Filling "LotFrontage" column with the mean of all values in this column exclude the missed ones
print("LotFrontage mean:" ,subset_hp_data.LotFrontage.dropna().mean())
subset_hp_data.LotFrontage.fillna(subset_hp_data.LotFrontage.dropna().mean())
```

```python
# Filling "Alley" column with the next value
hp_data.Alley.fillna(method='bfill', axis=0).fillna(0)[100:2000:100]
```
