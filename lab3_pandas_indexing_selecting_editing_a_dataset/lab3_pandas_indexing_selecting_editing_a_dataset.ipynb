### 1.7 Умовний вибір
Розглянемо вибір елементів набору даних, виходячи з умови

#### A) Проста умова
Наприклад, ми хочемо сформувати піднабір даних, який складається з книг, ціна яких є більшою за середню ціну.

```python
pmean=book_dat.Price.mean()
print("Середня ціна =",pmean)
```

```python
book_dat.loc[book_dat.Price>pmean] 
```

#### Б) Складна умова
Складна умова дозволяє виконувати логічні операції вибору з використанням логічних операторів «&» та «|».

```python
book_dat.loc[(book_dat.Price<pmean) & (book_dat.Year==2018)]
```

#### В) Умовні селектори 

«is in» – дозволяє вибирати дані, значення яких є у списку значень.
Наприклад, ось як ми можемо використовувати його для вибору книг, атором
яких є Jen Sincero.

```python
book_dat.loc[book_dat.Author.isin(['Jen Sincero'])]
```

Додаючи елементи до списку, можемо вибрати кілька елементів одночасно.
Наприклад, вибираємо публікації 2016 та 2017 років.

```python
book_dat.loc[book_dat.Year.isin([2016,2017])]
```

«isnull» (або «notnull»). Ці методи дозволяють виділити значення, які є
(або не є) порожніми (NaN).

```python
book_dat.loc[book_dat.Price.isnull()]
```

```python
book_dat.loc[book_dat.Year.isnull()]
```

#### 2.4 Умовний вибір

```python
#Виберемо дані, які нижчі за середню оцінку з математики
math_mean = score["math score"].mean()
print(math_mean)
score.loc[score["math score"] < math_mean]
```

```python
#Виберемо дані для жінок з результатом з математики нижче середнього
score.loc[(score["math score"] < math_mean) & (score.gender == "female")]
```

```python
#Виберемо дані для груп А та В
score.loc[score["race/ethnicity"].isin(['group B','group A'])]
```
