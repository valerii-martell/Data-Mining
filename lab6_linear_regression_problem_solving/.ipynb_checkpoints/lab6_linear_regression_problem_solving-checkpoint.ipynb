## 2. Practical part

### A simple dataset "Medical Cost Personal Datasets" was selected for this work from the following Kaggle topic:https://www.kaggle.com/mirichoi0218/insurance. 

### It contains the data about the US medical insurance cost for a humans according to their age, sex, bmi, number of children, region of residence and smoker status. Therefore this dataset is very suitable for Insurance Forecast by using Linear Regression.

***Links to download: [insurance.cvs](https://www.kaggle.com/mirichoi0218/insurance/download)***

### 2.1. Importing libraries and reading a dataset

```python
# Libraries importing
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Read a dataset
mydata = pd.read_csv("insurance.csv")
```

### 2.2. Data preprocessing

```python
# Show the top 5 rows of the datased we read
mydata.head()
```

```python
# Show the last 5 rows of the datased we read
mydata.tail()
```

```python
# Show the shape of the datased we read
mydata.shape
```

```python
# Get general information about the dataset after the removal
mydata.info()
```

```python
# Check the dataset for any missing data
mydata.isna().any()
```

```python
# Summarize the data to see the distribution of data
mydata.describe()
```

```python
# Encode the string data, 'sex' column, 'male' = 1, 'female' = 0
mydata.loc[mydata['sex'] == 'male', 'sex'] = 1
mydata.loc[mydata['sex'] == 'female', 'sex'] = 0

# Encode the string data, 'smoker' column, 'yes' = 1, 'no' = 0
mydata.loc[mydata['smoker'] == 'yes', 'smoker'] = 1
mydata.loc[mydata['smoker'] == 'no', 'smoker'] = 0

# Encode the string data, 'region' column, NE = 0, SE = 1, SW = 2, NW = 3
mydata.loc[mydata['region'] == 'northeast', 'region'] = 0
mydata.loc[mydata['region'] == 'southeast', 'region'] = 1
mydata.loc[mydata['region'] == 'southwest', 'region'] = 2
mydata.loc[mydata['region'] == 'northwest', 'region'] = 3
mydata
```

### 2.3. Division the dataset

```python
# Split the first 6 columns of the data
x,y = np.split(mydata,(6,),axis=1) 
```

```python
x = x.values
x
```

```python
y = y.values
y
```

```python
print('Sample data volume: %d, number of features: %d'%x.shape)
print('Target sample data volume:%d'%y.shape[0])
```

### 2.5. Model training

```python
# Importing libraries 
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
# Divide the data set into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Data normalization and standardization
ss = StandardScaler()
X_train = ss.fit_transform(x_train)
X_test = ss.transform(x_test)
```

**Polynomial regression** 

If the data is more complex than a simple straight line, a linear model can be used to fit the nonlinear data. A simple method is to add the power of each feature as a new feature, and then perform linear fitting on this extended feature set. This method is called polynomial regression.

```python
# Use Scikit-Learn's PolynomialFeatures to convert features
poly = PolynomialFeatures(2)
X_train = poly.fit_transform(X_train)
X_test = poly.fit_transform(X_test)
# Use Scikit-Learn's linear regression model 
lg = LinearRegression()
# Fit training data
lg.fit(X_train,y_train)
```

```python
# Calculate the mean square error of the training process
from sklearn.metrics import mean_squared_error
predictions = lg.predict(X_train)
mse = mean_squared_error(y_train, predictions)
rmse = np.sqrt(mse)
rmse
```

```python
# Check the r2 score
from sklearn.metrics import r2_score
r2_score=r2_score(y_train, predictions)
r2_score
```

```python
# Visualize the data fit of the training set
plt.figure(figsize=(16,10), facecolor='w')
ln_x_train = range(len(X_train))
# Raw data 
plt.plot(ln_x_train, y_train,'r-',lw=2,label=u'Real value')
# Pradictive values
plt.plot(ln_x_train,predictions,'b-',lw=2,label=u'Predictive value，$R^2$=%.3f'%r2_score)
# Show the result
plt.legend(loc='lower right')
plt.grid(True)
plt.title(u'The US medical insurance price prediction - training data')
plt.show()
```

```python
# Calculate the mean square error of the test process
from sklearn.metrics import mean_squared_error

test_pre = lg.predict(X_test)
mse = mean_squared_error(y_test, test_pre)
rmse = np.sqrt(mse)
rmse
```

```python
#Check the r2 score
from sklearn.metrics import r2_score
r2_score_pre=r2_score(y_test, test_pre)
r2_score_pre
```

```python
# Visualize the data fit of the training set
plt.figure(figsize=(16,10), facecolor='w')
ln_x_test = range(len(X_test))
# raw data 
plt.plot(ln_x_test, y_test,'r-',lw=2,label=u'real value')
# pradictive value
plt.plot(ln_x_test,test_pre,'b-',lw=2,label=u'Predictive value，$R^2$=%.3f'%r2_score_pre)
plt.legend(loc='lower right')
plt.grid(True)
plt.title(u'The US medical insurance price prediction - test data')
plt.show()
```
